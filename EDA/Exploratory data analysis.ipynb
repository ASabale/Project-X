```python
# Find missing values statistics

missing_value_map = {
    'prev_address_months_count': -1,
    'current_address_months_count': -1,
    'intended_balcon_amount': 'special',  # we'll handle this specially
    'bank_months_count': -1,
    'session_length_in_minutes': -1,
    'device_distinct_emails_8w': -1
}

missing_percent = {}

for col in df.columns:
    if col in missing_value_map:
        if col == 'intended_balcon_amount':
            # All negative values are missing for this feature
            n_missing = (df[col] < 0).sum()
        else:
            n_missing = (df[col] == missing_value_map[col]).sum()
    else:
        # For other columns, count NaNs if present
        n_missing = df[col].isna().sum()
    percent_missing = n_missing / len(df) * 100
    missing_percent[col] = percent_missing

# Convert to DataFrame for a nicer look
missing_df = pd.DataFrame.from_dict(missing_percent, orient='index', columns=['% Missing'])
missing_df = missing_df.sort_values('% Missing', ascending=False)

display(missing_df)
```

```python
# Prepare missing value criteria
missing_criteria = (
        (df['prev_address_months_count'] == -1) |
        (df['current_address_months_count'] == -1) |
        (df['intended_balcon_amount'] < 0) |
        (df['bank_months_count'] == -1) |
        (df['session_length_in_minutes'] == -1) |
        (df['device_distinct_emails_8w'] == -1)
)

df_clean = df[~missing_criteria].copy()
df_missing = df[missing_criteria].copy()

print("Rows after removing missing values:", df_clean.shape[0])
print("Rows with at least one missing value:", df_missing.shape[0])
```

```python
missing_conditions = {
    'prev_address_months_count': lambda x: x == -1,
    'current_address_months_count': lambda x: x == -1,
    'intended_balcon_amount': lambda x: x < 0,
    'bank_months_count': lambda x: x == -1,
    'session_length_in_minutes': lambda x: x == -1,
    'device_distinct_emails_8w': lambda x: x == -1
}

for col in features:
    # Select only rows where that feature is missing
    missing_rows = df[missing_conditions[col](df[col])]
    # You can show fraud distribution as a count or pie chart
    plt.figure(figsize=(5, 5))
    missing_rows[fraud_col].value_counts().plot.pie(autopct='%1.1f%%', labels=[f'Class {i}' for i in missing_rows[fraud_col].value_counts().index])
    plt.title(f'Fraud Distribution for Missing {col}')
    plt.savefig(f'Images/fraud distribution/Fraud Distribution for Missing {col} by Fraud Status percentage.png')
    plt.ylabel('')
    plt.show()
```

```python
def add_missing_indicators_and_impute(df):
    # Copy to avoid changing original
    new_df = df.copy()

    missing_value_map = {
        'prev_address_months_count': -1,
        'current_address_months_count': -1,
        'intended_balcon_amount': 'special',  # All negative values are missing
        'bank_months_count': -1,
        'session_length_in_minutes': -1,
        'device_distinct_emails_8w': -1
    }

    for col, missing_val in missing_value_map.items():
        # Create indicator
        if col == 'intended_balcon_amount':
            indicator = (new_df[col] < 0).astype(int)
            # Impute: use median of non-missing values
            median_val = new_df.loc[new_df[col] >= 0, col].median()
            new_df.loc[new_df[col] < 0, col] = median_val
        else:
            indicator = (new_df[col] == missing_val).astype(int)
            # Impute: use median of non-missing values
            median_val = new_df.loc[new_df[col] != missing_val, col].median()
            new_df.loc[new_df[col] == missing_val, col] = median_val

        new_df[col + '_missing'] = indicator

    return new_df

# Usage:
df_handled = add_missing_indicators_and_impute(df)
display(df_handled.head())
```

```python
categorical_cols = df_handled.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical features:", categorical_cols)
```

```python
# df_handled is your feature-engineered DataFrame
df_ohe = pd.get_dummies(df_handled, columns=categorical_cols, drop_first=False)

print(f"Shape after one-hot encoding: {df_ohe.shape}")
display(df_ohe.head())

# # Save the DataFrame with one-hot encoded features
# joblib.dump(df_ohe.columns.tolist(), 'ohe_columns.pkl')
# # new_event_encoded is your newly one-hot encoded DataFrame for an event
# ohe_columns = joblib.load('ohe_columns.pkl')
# new_event_encoded = new_event_encoded.reindex(columns=ohe_columns, fill_value=0)
```

```python
# Identify columns where all values are the same (i.e., only one unique value)
constant_cols = [col for col in df_ohe.columns if df_ohe[col].nunique() == 1]

print(f"Constant columns to be dropped: {constant_cols}")
```

```python
# Drop those columns
df_no_constant = df_ohe.drop(columns=constant_cols)

print(f"Shape after dropping constant columns: {df_no_constant.shape}")
display(df_no_constant.head())
```

```python
import xgboost as xgb

# Assume df_no_constant is your feature matrix, and 'fraud_bool' is your target
X = df_no_constant.drop('fraud_bool', axis=1)
y = df_no_constant['fraud_bool']

model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X, y)

# Get importance and sort
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)

display(feature_importance_df)
```

```python
os.makedirs('Images/Feature Importance', exist_ok=True)
plt.figure(figsize=(10, 10))
sns.barplot(
    data=feature_importance_df,
    y='Feature', x='Importance', palette='viridis'
)
plt.title(f'Top Feature Importances (XGBoost)')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig(f'Images/Feature Importance/Feature Importance after one hot coding.png')
plt.show()
```

```python
from imblearn.under_sampling import NearMiss, TomekLinks, EditedNearestNeighbours, ClusterCentroids

X = df_no_constant.drop('fraud_bool', axis=1)
y = df_no_constant['fraud_bool']

# 1. Tomek Links (removes borderline majority class points)
tl = TomekLinks(sampling_strategy='majority')
X_tl, y_tl = tl.fit_resample(X, y)
print(f"Tomek Links: {X_tl.shape}")

# 2. Edited Nearest Neighbours
enn = EditedNearestNeighbours()
X_enn, y_enn = enn.fit_resample(X, y)

# 3. NearMiss (example: NearMiss-1)
nm = NearMiss(version=1)
X_nm, y_nm = nm.fit_resample(X, y)

# 4. Cluster Centroids
cc = ClusterCentroids()
X_cc, y_cc = cc.fit_resample(X, y)

print(f"Tomek Links: {X_tl.shape}, ENN: {X_enn.shape}, NearMiss: {X_nm.shape}, ClusterCentroids: {X_cc.shape}")
```

```python
import matplotlib.pyplot as plt

def plot_class_counts(y, title="Class Distribution"):
    y.value_counts().plot(kind='bar')
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Count")
    plt.savefig(f'Images/Feature Importance/Feature Importance {title}.png')
    plt.show()

plot_class_counts(y, "Original Distribution")
plot_class_counts(y_tl, "After Tomek Links")
plot_class_counts(y_enn, "After ENN")
plot_class_counts(y_nm, "After NearMiss")
plot_class_counts(y_cc, "After Cluster Centroids")
```
