```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
# Load data
pd.set_option('display.max_columns', None)
df = pd.read_csv("data/data.csv") 
```

```python
display(df.head())        # first 5 rows
display(df.info())          # dtypes & non-null counts
display(df.describe(include='all'))  # summary stats for both numeric & categorical
```

```python
fraud_col = 'fraud_bool'

# Counts and percentages
counts = df[fraud_col].value_counts().sort_index()
percent = df[fraud_col].value_counts(normalize=True).sort_index() * 100
print("Counts:\n", counts)
print("\nPercentages:\n", percent.round(2))

# Pie chart
plt.figure(figsize=(5,5))
counts.plot.pie(autopct='%1.1f%%', labels=[f'Class {i}' for i in counts.index], startangle=90)
plt.ylabel('')
plt.title('Fraud Distribution')
plt.show()
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot a histogram for each numeric column
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    df[col].hist(bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np

# List of columns to log-transform
cols_to_transform = [
    'prev_address_months_count',
    'current_address_months_count',
    'days_since_request',
    'intended_balcon_amount',
    'zip_count_4w',
    'velocity_4w',
    'bank_branch_count_8w',
    'session_length_in_minutes'
]

# Create a new DataFrame to hold the transformed features
log_df = pd.DataFrame()

for col in cols_to_transform:
    # Compute offset so that the minimum value becomes 1 if there are non-positive values
    min_val = df[col].min()
    offset = 1 - min_val if min_val <= 0 else 0
    print(f"Transforming {col} with offset: {offset}")

    # Apply log transformation
    log_df[f'log_{col}'] = np.log(df[col] + offset)

# Preview the first few rows of the new DataFrame
log_df.head()
```

```python
# Plot histograms of original vs log-transformed features
for col in cols_to_transform:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    # Original
    axes[0].hist(df[col].dropna(), bins=30)
    axes[0].set_title(f'{col} (Original)')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')
    axes[0].grid(True)

    # Log-transformed
    axes[1].hist(log_df[f'log_{col}'].dropna(), bins=30)
    axes[1].set_title(f'log_{col} (Transformed)')
    axes[1].set_xlabel(f'log_{col}')
    axes[1].set_ylabel('Frequency')
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
cols = [
    'prev_address_months_count',
    'current_address_months_count',
    'days_since_request',
    'intended_balcon_amount',
    'zip_count_4w',
    'velocity_4w',
    'bank_branch_count_8w',
    'session_length_in_minutes'
]

# 1. Original DataFrame
df_original = df.copy()

# 2. Log-transformed DataFrame
df_log = df.copy()
for col in cols:
    df_log[col] = log_df[f'log_{col}']

# Preview shapes to confirm
print("Original shape:", df_original.shape)
print("Log-transformed shape:", df_log.shape)

# Optionally preview first few rows
display(df_original.head())
display(df_log.head())
```

```python
# Select numeric columns common to all DataFrames
numeric_cols = df_original.select_dtypes(include=['int64', 'float64']).columns

# Compute correlation matrices
corr_original = df_original[numeric_cols].corr()
corr_log = df_log[numeric_cols].corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_heatmap(corr_matrix, title):
    # Create a mask to display only the lower triangle of the matrix
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Set up the matplotlib figure
    plt.figure(figsize=(10, 8))

    # Draw the heatmap
    sns.heatmap(
        corr_matrix,
        mask=mask,
        cmap='coolwarm',
        square=True,
        linewidths=0.5,
        cbar_kws={"shrink": 0.5},
        xticklabels=True,
        yticklabels=True
    )

    # Set the title
    plt.title(title, fontsize=16)

    # Display the plot
    plt.show()

# Plot correlation matrices for original and log-transformed data
plot_correlation_heatmap(corr_original, 'Original Data Correlation Matrix')
plot_correlation_heatmap(corr_log, 'Log-Transformed Data Correlation Matrix')
```

```python
upper1 = corr_original.where(np.triu(np.ones(corr_original.shape), k=1).astype(bool))
upper2 = corr_log.where(np.triu(np.ones(corr_log.shape), k=1).astype(bool))

threshold = 0.55  # You can change this
to_drop1 = [column for column in upper1.columns if any(upper1[column] > threshold)]
to_drop2 = [column for column in upper2.columns if any(upper2[column] > threshold)]

print("Columns to drop:", to_drop1)
print("Columns to drop (log-transformed):", to_drop2)
```

```python
# after dropping the columns
df_reduced = df_original.drop(columns=to_drop1)
df_reduced_log = df_log.drop(columns=to_drop2)
```

```python
categorical_cols = df_reduced.select_dtypes(include='object').columns.tolist()

# Apply one-hot encoding to each dataset
df_original_encoded = pd.get_dummies(df_reduced, columns=categorical_cols, drop_first=False)
df_log_encoded = pd.get_dummies(df_reduced_log, columns=categorical_cols, drop_first=False)

# Display the first few rows of the encoded DataFrames
display(df_original_encoded.head())
display(df_log_encoded.head())
```

```python
from imblearn.over_sampling import SMOTE

def apply_smote(df, target_column):
    X = df.drop(columns=[target_column])
    y = df[target_column]

    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
    df_resampled[target_column] = y_resampled

    return df_resampled

target_column = 'fraud_bool'

df_original_smote = apply_smote(df_original_encoded, target_column)
df_log_smote = apply_smote(df_log_encoded, target_column)

display(df_original_smote.head())
display(df_log_smote.head())

df_original_smote.to_csv('data/original_encoded_smote.csv', index=False)
df_log_smote.to_csv('data/log_encoded_smote.csv', index=False)
```

```python
fraud_col = 'fraud_bool'

# Counts and percentages
counts = df_log_smote[fraud_col].value_counts().sort_index()
percent = df_log_smote[fraud_col].value_counts(normalize=True).sort_index() * 100
print("Counts:\n", counts)
print("\nPercentages:\n", percent.round(2))

# Pie chart
plt.figure(figsize=(5,5))
counts.plot.pie(autopct='%1.1f%%', labels=[f'Class {i}' for i in counts.index], startangle=90)
plt.ylabel('')
plt.title('Fraud Distribution')
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.features import FeatureImportances

# Prepare features and target
X = df_log_smote.drop(columns=['fraud_bool'])
X = pd.get_dummies(X)  # convert categorical variables
y = df_log_smote['fraud_bool']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Fit model and visualize feature importance
model = RandomForestClassifier()
viz = FeatureImportances(model)
viz.fit(X_train, y_train)
viz.show()
```
