```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
# Load data
pd.set_option('display.max_columns', None)
df = pd.read_csv("data/data.csv") 
```

```python
display(df.head())        # first 5 rows
display(df.info())          # dtypes & non-null counts
display(df.describe(include='all'))  # summary stats for both numeric & categorical
```

```python
missing = (
    df.isna().sum()
    .rename("n_missing")
    .to_frame()
    .assign(pct_missing=lambda x: 100 * x.n_missing / len(df))
    .sort_values("pct_missing", ascending=False)
)
display(missing)
```

```python
fraud_col = 'fraud_bool'

# Counts and percentages
counts = df[fraud_col].value_counts().sort_index()
percent = df[fraud_col].value_counts(normalize=True).sort_index() * 100
print("Counts:\n", counts)
print("\nPercentages:\n", percent.round(2))

# Pie chart
plt.figure(figsize=(5,5))
counts.plot.pie(autopct='%1.1f%%', labels=[f'Class {i}' for i in counts.index], startangle=90)
plt.ylabel('')
plt.title('Fraud Distribution')
plt.show()
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot a histogram for each numeric column
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    df[col].hist(bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer
import joblib

# List of columns to transform
cols_to_transform = [
    'prev_address_months_count',
    'current_address_months_count',
    'days_since_request',
    'intended_balcon_amount',
    'zip_count_4w',
    'velocity_4w',
    'bank_branch_count_8w',
    'session_length_in_minutes'
]

# Initialize the PowerTransformer with Yeo-Johnson method
pt = PowerTransformer(method='yeo-johnson', standardize=False)

# Apply the transformation
df_yeojohnson = df.copy()
df_yeojohnson[cols_to_transform] = pt.fit_transform(df[cols_to_transform])

# Save
joblib.dump(pt, 'data/yeo_johnson_transformer.joblib')

# Plot original vs. transformed distributions side by side
for col in cols_to_transform:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Original distribution
    axes[0].hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[0].set_title(f'{col} (Original)')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')

    # Transformed distribution
    axes[1].hist(df_yeojohnson[col].dropna(), bins=30, color='lightgreen', edgecolor='black')
    axes[1].set_title(f'{col} (Yeo-Johnson Transformed)')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Preview the first few rows of the new DataFrame
df_yeojohnson.head()
```

```python
# Select numeric columns common to all DataFrames
numeric_cols = df_yeojohnson.select_dtypes(include=['int64', 'float64']).columns

# Compute correlation matrices
corr_original = df_yeojohnson[numeric_cols].corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_heatmap(corr_matrix, title):
    # Create a mask to display only the lower triangle of the matrix
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Set up the matplotlib figure
    plt.figure(figsize=(10, 8))

    # Draw the heatmap
    sns.heatmap(
        corr_matrix,
        mask=mask,
        cmap='coolwarm',
        square=True,
        linewidths=0.5,
        cbar_kws={"shrink": 0.5},
        xticklabels=True,
        yticklabels=True
    )

    # Set the title
    plt.title(title, fontsize=16)

    # Display the plot
    plt.show()

# Plot correlation matrix
plot_correlation_heatmap(corr_original, 'Transformed Data Correlation Matrix')
```

```python
upper = corr_original.where(np.triu(np.ones(corr_original.shape), k=1).astype(bool))

threshold = 0.55  # You can change this
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print("Columns to drop:", to_drop)
```

```python
# after dropping the columns
df_reduced = df_yeojohnson.drop(columns=to_drop)
```

```python
categorical_cols = df_reduced.select_dtypes(include='object').columns.tolist()

# Apply one-hot encoding to each dataset
df_encoded = pd.get_dummies(df_reduced, columns=categorical_cols, drop_first=False)

# Display the first few rows of the encoded DataFrames
display(df_encoded.head())
```

```python
from imblearn.over_sampling import SMOTE

def apply_smote(df, target_column):
    X = df.drop(columns=[target_column])
    y = df[target_column]

    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
    df_resampled[target_column] = y_resampled

    return df_resampled

target_column = 'fraud_bool'

df_smote = apply_smote(df_encoded, target_column)

display(df_smote.head())

df_smote.to_csv('data/encoded_smote.csv', index=False)
```

```python
fraud_col = 'fraud_bool'

# Counts and percentages
counts = df_smote[fraud_col].value_counts().sort_index()
percent = df_smote[fraud_col].value_counts(normalize=True).sort_index() * 100
print("Counts:\n", counts)
print("\nPercentages:\n", percent.round(2))

# Pie chart
plt.figure(figsize=(5,5))
counts.plot.pie(autopct='%1.1f%%', labels=[f'Class {i}' for i in counts.index], startangle=90)
plt.ylabel('')
plt.title('Fraud Distribution')
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.features import FeatureImportances

# Prepare features and target
X = df_smote.drop(columns=['fraud_bool'])
X = pd.get_dummies(X)  # convert categorical variables
y = df_smote['fraud_bool']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Fit model and visualize feature importance
model = RandomForestClassifier()

# Fit the model and make visualization
visualizer = FeatureImportances(model, relative=False, sort=True)
visualizer.fit(X_train, y_train)
visualizer.show(outpath='data/feature_importances.png')
```
